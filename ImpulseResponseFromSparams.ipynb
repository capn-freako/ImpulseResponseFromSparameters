## Introduction

### Historical Context

On August 25, 2020 I had [an article published in _Signal Integrity Journal_ (SIJ) titled, "Impulse Response from Insertion Loss"](https://www.signalintegrityjournal.com/articles/1847-impulse-response-from-insertion-loss)
In my article I explained how the raw, positive frequency only, data contained in a _Touchstone_&reg; file needed to be conjugate duplicated and reordered, in order to get a correct impulse response from a typical `ifft()` function.
I also went on to explain how applying _windowing_ to the data, before calling `ifft()`, can help eliminate the non-causal ringing, which typically occurs as a result of the incomplete data in most Touchstone files.

Later, on May 12, 2022, Professors Morales and Agili, of Penn State Harrisburg's _Center for Signal Integrity_, presented a SIJ webinar titled, "The Use of DSP Techniques to Analyze S-Parameter Sampled Data."
Their webinar was partly a rebuttal to my own previous article.
In fact, they used some of my figures unaltered, presumably to make it quite clear that they were rebutting my suggestion of the use of windowing techniques to smooth non-causal ringing.
In their webinar, they propose, as an alternative to windowing, the use of certain DSP techniques, in particular the [_Bilinear Transform_](https://en.wikipedia.org/wiki/Bilinear_transform), to better transform the raw frequency domain S-parameter data in a Touchstone file into the _discrete_ time domain, for further processing there.

### Statement of Intent

While Morales and Agili make an excellent contribution to our collective understanding of the issues involved with bringing band limited frequency domain data into a discrete time signal processing flow for channel modeling and simulation, their work suffers from several ailments:

1. Their work is a bit unclear, incomplete, and in certain sections incorrect, potentially misleading novice readers.
Additionally, their final example, regarding accurate temporal location of small reflections in the channel impulse response, is incorrect, as I'll show here.  
I'll try to do two things here:

    1. add clarity, completeness, and necessary corrections to their original arguments, and
    
    1. test their suggested approach on the actual Touchstone data used in my original article.
    
1. They present their chosen approach as always better than the windowing approach I'd previously advocated.
That is, they imply that their DSP based approach, invoking the _Bilinear Transform_, is _always_ better than a windowing based approach.
And this is simply not true, based on my findings here.

## Clarification of Some Concepts in [Morales and Agili]

Here, I'll try to make some corrections, fill in some missing detail, or offer a complimentary perspective on the previous work of Morales and Agili, by slide number.

### Slide 7

Their statement:

> S-parameter data of analog high-speed channels are based on Laplace transform i.e., in the continuous frequency domain.

is misleading.
S-parameter data isn't based on any type of transform.
It is simply the measured frequency response of a certain path through the structure: insertion, return, NEXT, or FEXT.
And, at least as typically delivered in Touchstone files, S-parameter data is _not_ continuous, but rather _sampled_, or _discrete_.
They actually concede this, in their following bullet point, but hedge their concession with, "This fact does not mean that the data are coming from digital systems."
Well, of course the S-parameter data in Touchstone files is not coming from a digital system.
It is coming from measurements made using a VNA; isn't that obvious?


### Slide 9

1. I have no idea why they chose to make their drawing and associated equation so unnecessarily complicated, but let's just state the well known (and much simpler) correct equation describing the input/output relationship of a 2-port _linear time invariant_ (LTI) system, for the record:

    $$
    S_{2,1}(f) = \frac{V^-_2(f)}{V^+_1(f)} \cdot \sqrt{\frac{Z_1(f)}{Z_2(f)}}
    $$

    or:

    $$
    V^-_2(f) = S_{2,1}(f) \cdot V^+_1(f) \cdot \sqrt{\frac{Z_2(f)}{Z_1(f)}}
    $$

    where:
    
    - $V^-_2(f)$ is the (complex) amplitude of the sinusoidal voltage wave exiting from port #2 at frequency, $f$.
    - $V^+_1(f)$ is the (complex) amplitude of the sinusoidal voltage wave entering into port #1 at frequency, $f$.
    - $Z_1(f)$ and $Z_2(f)$ are the (complex) impedances of ports 1 and 2, respectively, at frequency, $f$.
    
1. Their statement:

    > The interaction between the source voltage and the DUT, produces frequency responses of the analog system in the form of S- parameters for each individual sinusoidal frequencies, provided that the system reaches a steady state.
    
    is erroneous, as follows:
    
    1. "...interaction between the source..., produces frequency responses..."
    
        The frequency response of a LTI system is _not_ "produced" by anything external to it.
        It is an implicit property of that LTI system.
        
    1. "...in the form of S-parameters..."
    
        S-parameters are not some universal, or fundamental, "form" of the frequency response of a LTI system.
        They are a standardized convention for _tabulating_ the frequency response of a LTI system.
        We could easily choose a _different_ convention for tabulating the frequency response.
        If we did then only the _convention_ changes; the frequency response remains exactly the same.
        
    1. "...provided that the system reaches a steady state."
    
        The frequency response of a LTI system exists _always_!
        It is a fundamental property of such a system.
        It is _not_ predicated upon the system being driven such that its output reaches some steady state.
        

### Slide 10

They're claiming steady state is reached at a time, $5 \tau$, as if that time were a property of the system they're simulating, but it's not.
It's merely an artifact of their chosen length of signal vector.
Observe what happens when we try 3 different signal vector lengths for the same simulation:

````python
from matplotlib import pyplot as plt
import numpy as np

a    = 1
for tmax in [5, 10, 20]:
    print(f"tmax = {tmax}:")
    t    = np.linspace(-1,tmax,(tmax+1)*100)
    Vs   = np.cos(2*np.pi*t)
    u    = np.zeros(t.shape)
    u[np.where(t >= 0)] = 1
    hDUT = np.exp(-a*t) * u
    y    = np.convolve(Vs, hDUT, mode="same")

    plt.figure(figsize=(10,4))
    plt.subplot(121)
    plt.plot(t,Vs,  'r',label="Vs")
    plt.plot(t,hDUT,'b',label="hDUT")
    plt.xlabel("Time (norm.)")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.legend()
    plt.subplot(122)
    plt.plot(t,y,'b')
    plt.title("Convolution of Vs and hDUT")
    plt.xlabel("Time (norm.)")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.show()
````

Note the decay towards "steady state" does not begin at some time, $\tau$, which is a property of the network being simulated, but simply when the defined input signal test vector expires!

### Slide 11

The whole underlying premise of this slide seems to be that some time dependent decay, or "steady state" achievement, is fundamental to the measurement of a network frequency response.
That is completely wrong.
The measurement of the frequency response of a network involves two very simple steps:

1. Stimulate the input port of the network with a _continuous_ sinusoidal voltage of a particular frequency through a matched impedance.

1. Measure the resultant magnitude and phase of the sinusoidal voltage coming out of the output port of the network into a matched load.

That's it. There's nothing more.
This expressed notion of having to wait until steady state has been achieved before "launching the next sinusoidal stimulus" betrays a complete misunderstanding on the part of Morales and Agili about what it means to measure the frequency response of a network.

### Slide 12

It's difficult to divine the intended main point of this slide, but that last equation is suggestive that what Morales and Agili are trying to explain is the classic outcome of sampling theory: that the spectrum of the sampled signal repeats at intervals of: $f_r = \frac{1}{t_s}$, where $t_s$ is the sampling interval.

The problem is we're going in the opposite direction in this work.
That is, we're _starting_ with sampled _frequency_ domain data, namely the tabulated S-parameters in the Touchstone file.
That's okay, though, because this relationship is bidirectional (perhaps, what they're referring to in the slide title?).
That is, sampling in the frequency domain with sampling interval: $f_s$, will lead to repetition of the reconstructed time domain signal, at intervals: $t_r = \frac{1}{2 \pi f_s}$.

Let's assume that that's what they meant to say and see where it leads us...

A very typical sampling interval w/ Touchstone data meant for high speed channel characterization is 10 MHz.
At that frequency sampling interval, what is our time domain repetition interval?
$t_r = \frac{1}{2 \pi \cdot 10MHz}$ = 16 ns.
I'm pretty sure we're going to loose interest in our reconstructed channel impulse response after a few hundred picoseconds.
So, we appear to have a couple orders of magnitude of "I don't care" margin.

### System Vector Generation

It's very easy, when doing this work, to end up with time and frequency vectors that are out of whack with each other.
So, we'll encapsulate the generation of our "system" vectors in a function.
That way, if we make a mistake in the coding of these vectors then it will only need to be repaired once.

````python
def make_vecs_norm(Npts):
    """Construct system vectors, using normalized time/frequency.
    
    Args:
        Npts (int): # of points in returned vectors.
            (Should be integral power of 2.)
    """
    t    = np.arange(Npts)
    f0   = 1/Npts                            # Fundamental frequency
    fp   = [n * f0 for n in range(Npts//2)]  # Starts w/ "0" and doesn't include Nyquist.
    fn   = list(map(lambda x: -x, fp[1:]))   # Construct the negative frequencies,
    fn.reverse()                             # in reverse order,
    f    = np.array(fp + [f0*Npts/2] + fn)   # as discussed in my original article.
    fp   = np.array(fp)                      # For plotting utility.
    w    = 2*np.pi*f
    s    = 1j*w
    z    = np.exp(s)
    return (t, f, w, s, z, fp)

Npts = 8192  # Nice compromise between computation time and plot smoothness.
t, f, w, s, z, fp = make_vecs_norm(Npts)

# Plot things, as a sanity check on our constructions.
plt.figure(figsize=(10,4))
plt.plot(f)
plt.title("Values in System Frequency Vector (norm.)")
plt.show()

plt.figure(figsize=(10,4))
plt.subplot(121)
plt.plot(z.real)
plt.title("Z - Real Part")
plt.subplot(122)
plt.plot(z.imag)
plt.title("Z - Imaginary Part")
plt.show()
````

### Channel Frequency Response Creation

Encapsulating coding logic in functions is such a good practice, in general, that we'll use it again to generate their original channel frequency response specifications.

````python
def calc_chnl_H_orig(s, causal=True):
    """Calculate their original channel frequency response
    for a particular s-vector.
    
    Args:
        s ([complex(float)]): s-values for which to calculate response.
        
    KeywordArgs:
        causal (bool): Set to `False` for non-causal response.
            Default: True.
            
    Returns:
        [complex(float)]: The (complex) channel frequency response
            for the given `s` values.
    """
    if causal:
        a = -1
    else:
        a =  1
    return 0.98/(s+1) + 0.04*np.exp(a*6*s)/(s+2)

H1 = calc_chnl_H_orig(s)
H2 = calc_chnl_H_orig(s, causal=False)

plt.figure(figsize=(10,4))
plt.subplot(121)
plt.plot(fp, 20*np.log10(np.abs(H1[:len(fp)])), label="causal")
plt.plot(fp, 20*np.log10(np.abs(H2[:len(fp)])), label="non-causal")
plt.title("Frequency Response - Magnitude")
plt.xlabel("Frequency (norm.)")
plt.ylabel("Mag.(H) (dB)")
plt.grid()
plt.legend()

plt.subplot(122)
plt.plot(fp, np.unwrap(np.angle(H1[:len(fp)]))*180/np.pi, label="causal")
plt.plot(fp, np.unwrap(np.angle(H2[:len(fp)]))*180/np.pi, label="non-causal")
plt.title("Frequency Response - Phase")
plt.xlabel("Frequency (norm.)")
plt.ylabel("Phase(H) (deg.)")
plt.grid()
plt.legend()

plt.show()
````

Both plots above look correct.
Our main pole is at: $\omega = 1$, which corresponds to: $f = \frac{1}{2 \pi} \approx 0.16$.
And that is about where we see both the -3dB magnitude and 45 degree phase response.

### Channel Impulse Response

Now, let's finally take the `IFFT()` of our channel frequency responses, as they prescribe, and see if things match.  
We'll repeat the procedure for several different values for `Npts`, to assess the effect of that parameter on our output.

````python
for Npts in [8, 16, 32, 64, 128]:
    print(f"Npts = {Npts}")
    t, f, w, s, z, fp = make_vecs_norm(Npts)
    H1 = calc_chnl_H_orig(s)
    H2 = calc_chnl_H_orig(s, causal=False)
    h1 = np.fft.ifft(H1)
    h2 = np.fft.ifft(H2)
    t_shift = t - max(t)//2
    plt.figure(figsize=(10,4))
    plt.plot(t_shift, np.fft.fftshift(h1), label="causal")
    plt.plot(t_shift, np.fft.fftshift(h2), label="non-causal")
    plt.title("Original Channel Impulse Responses")
    plt.axis(xmin=-50, xmax=50)
    plt.legend()
    plt.show()
````

We observe the following:

1. Our choice of `Npts` appears to have no effect on the final output, other than limiting the extent of our view.

1. The differences between the "causal" and "non-causal" channel impulse responses in our plots, above, are much less severe than what they show on their slide 13.

We may need to ask Professors Morales and Agili to share their code with us, so we can see where we are diverging from their approach.

## Recreating Slide 18

The plot first shown on slide 18 is their coup de grace.
It is intended to clearly demonstrate the superiority of their _Bilinear Transform_ based approach.
Let's dig into it a little bit and see what we can learn...

### Bilinear Transform Application

On their slide 18, Professors Morales and Agili advise us to apply the _Bilinear Transform_, as follows:

1. Modify our s-vector according to:

    $$
    s = \frac{2}{T_s} \frac{z-1}{z+1}
    $$
    
1. Recalculate our channel frequency responses, using this new s-vector.

1. Take the `IFFT()` of those new frequency responses, to get more accurate channel impulse responses.

(Note that the flow above is reinforced by the flowchart on their slide 20.)

Let's try this...

````python
Npts = 8192  # Just to make sure that we don't miss anything, due to insufficient resolution.
print(f"Npts = {Npts}")
t, f, w, s, z, fp = make_vecs_norm(Npts)
sz = 2 * (z-1)/(z+1)  # `Ts` = 1 when working in normalized time.

H1 = calc_chnl_H_orig(sz)
H2 = calc_chnl_H_orig(sz, causal=False)
h1 = np.fft.ifft(H1)
h2 = np.fft.ifft(H2)
t_shift = t - max(t)//2

plt.figure(figsize=(10,4))
plt.plot(t_shift, np.fft.fftshift(h1), label="causal")
plt.plot(t_shift, np.fft.fftshift(h2), label="non-causal")
plt.title("Original Channel Impulse Responses")
plt.axis(xmin=-50, xmax=50)
plt.legend()
plt.show()
````

Okay, our plot isn't an exact match to theirs on slide 18, but the general features are correct:
the reflective bump occurs _before_ time zero in the non-causal case and _after_ time zero in the causal case.
And I think that was the main point of their plot.

### An Inconvenient Truth

Despite the reassuring nature of the plot above, it is wrong.
The method prescribed for application of the _Bilinear Transform_ (BT) by professors Morales and Agili is incorrect.
The reason is simple: the `ifft()` function has very explicit expectations, regarding the frequencies that are associated with each element of its input vector.
(This was the main point of my original article.)
And we have violated those expectations by "warping" the frequency values, via the BT.
The fact that the plot above "looks" correct is merely coincidental.
To support this claim, I'll do two things:

1. I'll send an obviously incorrect vector into the `ifft()` function and show that it, also, produces output that "looks" correct.

1. Later, when we've moved into absolute frequency space, I'll show that application of the BT as they prescribe yields clearly nonsensical results.

For now, here is item 1:
The most common mistake made by novice engineers when using the `ifft()` function is to leave out the conjugate response elements, corresponding to the negative frequencies, from the input vector.
Let's make that mistake and see what happens...

````python
t, f, w, s, z, fp = make_vecs_norm(Npts)
H1 = calc_chnl_H_orig(s)
H2 = calc_chnl_H_orig(s, causal=False)
h1 = np.fft.ifft(H1[:Npts//2])  # Omitting the conjugate response elements.
h2 = np.fft.ifft(H2[:Npts//2])
plt.figure(figsize=(10,4))
plt.plot(np.fft.fftshift(h1), label="causal")
plt.plot(np.fft.fftshift(h2), label="non-causal")
plt.title("Original Channel Impulse Responses")
plt.axis(xmin=2020, xmax=2080)
plt.legend()
plt.show()
````

Note that, again, we see correctly positioned reflective "bumps" for both the causal and non-causal case.
And, were our mistake not so obvious, we might conclude that our analysis was correct.

### A More Physically Correct Model

Astute readers probably noticed that the channel specifications offered by Professors Morales and Agili are a bit suspect.
And they may be wondering now how making those specifications more physically correct might affect the above analysis.
So, before moving on, let's address this.

Let's pause a moment to consider the nature of their channel specifications:

$$
\begin{eqnarray}
H(s)   &=& \frac{0.98}{s + 1} + \frac{0.04 e^{-6s}}{s + 2} \qquad \text{(causal)} \\
H_1(s) &=& \frac{0.98}{s + 1} + \frac{0.04 e^{+6s}}{s + 2} \qquad \text{(non-causal)}
\end{eqnarray}
$$

What does such a channel look like physically?

Well, the first term is easy.
It describes a single pole low pass filter with cutoff frequency: 1 rad./sec., and a _transmission coefficient_, $T$, of 98%.

The second term, however, is a bit odd.
It, also, represents a single pole low pass filter, but its cutoff frequency is twice that of the first term.  
(That's why a "0.04" was required, as opposed to the more intuitively suggestive "0.02".)  
Why? I suspect that there is no valid _physical_ reason.

The exponential term in the numerator: $e^{-6s}$, describes a phase _delay_
($e^{+6s}$ describes an _advance_, which is what makes the second case non-causal.),
which is _linearly proportional_ to frequency.  
In the time domain that is just a simple _propagation delay_.  
So, the second term represents a _reflection_ at the (imperfectly matched) Rx input termination.

That reflection travels back through the channel described by the first term, gets completely re-reflected (with amplitude inversion) by the (presumably, perfect zero Ohm) driver, and travels through the same channel one last time, before being tallied by the sum.  
(Negative reflections at _both_ the Rx and Tx are why the "0.04" is positive.)  
(Note that the "6" represents the _round trip_ delay.
So, the channel itself has a propagation delay of 3, in normalized time units.)

So, if that second term represents a reflection making a round trip through the same channel described by the first term then why does it have a cutoff frequency twice that of the first term?  
I don't know, but strongly suspect that a more physically correct form for the second term in the sum is:

$$
\frac{0.02 e^{-6s}}{(s + 1)^2}
$$

What happens if we repeat their prescribed application of the _Bilinear Transform_ using this (presumably) more physically correct expression?

````python
def calc_chnl_H_mod(s, causal=True):
    """Calculate modified channel frequency response
    for a particular s-vector.
    
    Args:
        s ([complex(float)]): s-values for which to calculate response.
        
    KeywordArgs:
        causal (bool): Set to `False` for non-causal response.
            Default: True.
            
    Returns:
        [complex(float)]: The (complex) channel frequency response
            for the given `s` values.
    """
    if causal:
        a = -1
    else:
        a =  1
    return 0.98/(s+1) + 0.02*np.exp(a*6*s)/(s+1)**2

Npts = 8192
t, f, w, s, z, fp = make_vecs_norm(Npts)

H1 = calc_chnl_H_mod(s)
H2 = calc_chnl_H_mod(s, causal=False)

plt.figure(figsize=(10,4))
plt.subplot(121)
plt.plot(fp, 20*np.log10(np.abs(H1[:len(fp)])), label="causal")
plt.plot(fp, 20*np.log10(np.abs(H2[:len(fp)])), label="non-causal")
plt.title("Frequency Response - Magnitude")
plt.xlabel("Frequency (norm.)")
plt.ylabel("Mag.(H) (dB)")
plt.grid()
plt.legend()

plt.subplot(122)
plt.plot(fp, np.unwrap(np.angle(H1[:len(fp)]))*180/np.pi, label="causal")
plt.plot(fp, np.unwrap(np.angle(H2[:len(fp)]))*180/np.pi, label="non-causal")
plt.title("Frequency Response - Phase")
plt.xlabel("Frequency (norm.)")
plt.ylabel("Phase(H) (deg.)")
plt.grid()
plt.legend()

plt.show()
````

The plots above look very similar to those of Sec. 3.2, but with much less "wiggling".

Continuing on...

````python
h1 = np.fft.ifft(H1)
h2 = np.fft.ifft(H2)
t_shift = t - max(t)//2
plt.figure(figsize=(10,4))
plt.plot(t_shift, np.fft.fftshift(h1), label="causal")
plt.plot(t_shift, np.fft.fftshift(h2), label="non-causal")
plt.title("Modified Channel Impulse Responses")
plt.axis(xmin=-50, xmax=50)
plt.legend()
plt.show()
````

## Moving to Absolute Frequency

### Choosing a Sampling Interval

In order to create our $z$ vector, we have to choose a sampling interval, $T$, as: $z = e^{sT}$.
How to choose $T$?
What criteria are available to guide our choice here?
Well, we know that we have valid frequency domain data in the Touchstone files only out to 35 GHz (because that's where we see the plots suddenly drop to zero).
And we know that, in order to prevent aliasing below 35 GHz, we'll have to be sampling at, at least, 70 GHz.
So, let's use 100 GHz as our sampling rate.
That yields: $T = \frac{1}{100 GHz} = 10 ps$.


````python
def make_vecs(Ts, Npts):
    """Construct system vectors.
    
    Args:
        Ts (float): sampling interval (s).        
        Npts (int): # of points in returned vectors.
            (Should be integral power of 2.)
    """
    t    = np.array([n * Ts for n in range(Npts)])
    f0   = 1/(Ts * Npts)                     # Fundamental frequency
    fp   = [n * f0 for n in range(Npts//2)]  # Starts w/ "0" and doesn't include Nyquist.
    fn   = list(map(lambda x: -x, fp[1:]))   # Construct the negative frequencies,
    fn.reverse()                             # in reverse order,
    f    = np.array(fp + [f0*Npts/2] + fn)   # as discussed in my original article.
    fp   = np.array(fp)  # For plotting utility.
    w    = 2*np.pi*f
    s    = 1j*w
    z    = np.exp(s * Ts)
    return (t, f, w, s, z, fp)

Ts   = 10e-12  # As per discussion above.
Npts = 8192    # Nice compromise between computation time and plot smoothness.
t, f, w, s, z, fp = make_vecs(Ts, Npts)

# Plot things, as a sanity check on our constructions.
plt.figure(figsize=(10,4))
plt.plot(f/1e9)
plt.title("Values in System Frequency Vector (GHz)")
plt.show()

plt.figure(figsize=(10,4))
plt.subplot(121)
plt.plot(z.real)
plt.title("Z - Real Part")
plt.subplot(122)
plt.plot(z.imag)
plt.title("Z - Imaginary Part")
plt.show()
````

### Channel s-Frequency Response

They don't show us what these two channels look like in the frequency domain, before and after applying the _Bilinear Transform_.
Let's take a look for ourselves...

**Note:** At this point we have a problem, because Morales and Agili are working in _normalized_ time/frequency and don't tell us their normalization constant.
So, we have to guess.
But, how do we do that?

We start by guessing the bandwidth of the channel they're describing, as this tells us what the "1" in the denominator of the left term in their sums is equivalent to.
I'm just going to guess that it has the same bandwidth as my original Touchstone data: about 20 GHz.

Next, we note that they've expressed the _phase delay_ of a _round trip_ through their channel as: $e^{-6s} = e^{-j 2 \pi \cdot 6 f'}$, where $f'$ is normalized frequency in units of the channel bandwidth, which I've guessed to be: 20 GHz.
That means that a _single pass_ through their channel represents one cycle of delay at:

$$
\begin{eqnarray}
3 f'_{2 \pi} &=& 1 \\
f'_{2 \pi} &=& \frac{1}{3} \\
f_{2 \pi} &=& \frac{1}{3} \cdot 20 GHz = 6.667 GHz
\end{eqnarray}
$$

This gives us a _channel propagation delay_ of: $\tau_d = \frac{3}{20 GHz} = 150 ps$  
and, assuming a _propagation velocity_: $\nu = \frac{1}{2} c$, a physical channel length of: $l = \nu \tau_d = 22.5 mm$, which falls right in the middle of the typical length range for XSR links: 10 - 50 mm.

Recasting their frequency domain channel expressions, using both our correction to their second term and our guess as to their chosen normalization constant:

$$
\begin{eqnarray}
H(s)   &=& K \frac{0.98}{s + 2 \pi \cdot 20 GHz} + K^2 \frac{0.02 e^{-6E^{-10} j \pi f}}{(s + 2 \pi \cdot 20 GHz)^2} \qquad \text{(causal)} \\
H_1(s) &=& K \frac{0.98}{s + 2 \pi \cdot 20 GHz} + K^2 \frac{0.02 e^{+6E^{-10} j \pi f}}{(s + 2 \pi \cdot 20 GHz)^2} \qquad \text{(non-causal)}
\end{eqnarray}
$$

where $f$ now has units of Hz.

````python
def calc_chnl_H(s, td=150e-12, fc=20e9, R=-0.02, causal=True):
    """Calculate channel frequency response for a particular s-vector.
    
    Note the slight mathematical change, relative to the original
    prescription of Morales and Agili, to be more physically correct.
    
    Args:
        s ([complex(float)]): Array of complex floats.
        
    KeywordArgs:
        td (float): Single pass time delay of channel (s).
            Default: 150 ps.
        fc (float): Low pass filter corner frequency (Hz).
            Default: 20 GHz.
        R (float): Reflection coefficient looking into Rx term.
            Default: -2%.
        causal (bool): Set to `False` for non-causal response.
            Default: True.
            
    Returns:
        [complex(float)]: The (complex) channel frequency response
            for the given `s` values.
        
    Notes:
        1. Channel is modeled as single pole low pass filter.
        2. Complete return of reflected energy by driver is assumed.
        3. Only one round-trip reflection is tallied.
    """
    if causal:
        a = -1
    else:
        a =  1
    p  = -2 * np.pi * fc
    H0 = -p * (1 + R) / (s - p) * np.exp(a*s*td)  # single pass response
    return H0 - R * H0**2

H  = calc_chnl_H(s)
H1 = calc_chnl_H(s, causal=False)

plt.figure(figsize=(10,4))
plt.subplot(121)
plt.semilogx(fp/1e9, 20*np.log10(np.abs(H[:len(fp)])),  label="causal")
plt.semilogx(fp/1e9, 20*np.log10(np.abs(H1[:len(fp)])), label="non-causal")
plt.title("Frequency Response - Magnitude")
plt.xlabel("Frequency (GHz)")
plt.ylabel("Mag.(H) (dB)")
plt.grid()
plt.legend()

plt.subplot(122)
plt.semilogx(fp/1e9, np.unwrap(np.angle(H[:len(fp)]))*180/np.pi,  label="H")
plt.semilogx(fp/1e9, np.unwrap(np.angle(H1[:len(fp)]))*180/np.pi, label="H1")
plt.title("Frequency Response - Phase")
plt.xlabel("Frequency (GHz)")
plt.ylabel("Phase(H) (deg.)")
plt.grid()
plt.legend()

plt.show()
````

And, with the exception of the "wiggles" caused by the small reflective term, we see what looks like a typical single pole response, with -3dB magnitude occurring at our guessed bandwidth of 20 GHz.

## Proper Use of Bilinear Transformation

So, if the prescription given by Professors Morales and Agili for applying the _Bilinear Transform_ (BT) is incorrect then how do we correctly apply the BT?

We follow these steps:

1. Shift all poles and zeros in our system specification, according to the frequency warping invoked by the BT.

1. Rearrange our thusly modified system transfer function, such that it is in _standard normalized expanded_ form, which implies:

    1. It is expressed as the product of a constant gain factor and a ratio of two polynomials in $z$, such that:
    
    1. The numerator has degree less than or equal to the denominator.
    
    1. Both numerator and denominator have been multiplied by: $z^k$, such that the highest power of $z$ in the denominator is zero.
    
    1. Both numerator and denominator have been normalized by the same constant factor, such that the coefficient of the $z^0$ term in the denominator is one.
    
<div alert class="alert-info">
Note that, at this point, we could stop and assess the unit impulse respponse of our digital system "manually", because when expressed in this form the transfer function of a discrete time system happens also to correspond to its standard difference equation, i.e., its <em>time domain</em> behavior.<p>
(Why this oh so fundamental feature of DSP was omitted in Professors Morales and Agili's presentation is beyond me.)<p>
However, doing so might arouse suspicion.
    So, instead, let's make use of the functions provided to us in the <em>NumPy/SciPy</em> libraries.
We'll avoid mistakes, as well as any perception of impropriety.
</div>

1. Use the `dimpulse()` function to view the equivalent _unit impulse response_ of our digital system.

### Transforming the Channel Specification

Taking our modification to their original channel specification as an example:

$$
\begin{eqnarray}
H(s)   &=& \frac{-0.98 p}{s - p} + \frac{0.02 e^{-2 \tau_d s} p^2}{(s - p)^2} \\
       &=& -p \frac{0.98 s - p \left( 0.98 + 0.02 e^{-2 \tau_d s} \right)}{s^2 - 2 p s + p^2}
\end{eqnarray}
$$

where, as previously determined:

$$
\begin{eqnarray}
p      &=& -2 \pi \cdot 20 GHz \\
\tau_d &=& 150 ps
\end{eqnarray}
$$

Then, substituting in for $z$ according to the BT:

$$
\begin{eqnarray}
H(z) &=& H(s) |_{s = \frac{2}{T} \frac{z - 1}{z + 1}} \\
     &=& -p \frac{0.98 \frac{2}{T} \frac{z - 1}{z + 1} -p \left( 0.98 + 0.02 z^{-2 \frac{\tau_d}{T}} \right) }{\left( \frac{2}{T} \frac{z - 1}{z + 1} \right)^2 - 2 p \frac{2}{T} \frac{z - 1}{z + 1} + p^2}
\end{eqnarray}
$$

where we've used the equality: $z = e^{sT}$, to simplify the reflection term.

Now, multiplying numerator and denominator by: $\frac{T^2}{4} (z+1)^2$:

$$
H(z) = -p \frac{0.98 \frac{T}{2} (z+1) (z-1) -p \frac{T^2}{4} (z+1)^2 \left( 0.98 + 0.02 z^{-2 \frac{\tau_d}{T}} \right) }{(z-1)^2 - 2 p \frac{T}{2} (z+1) (z-1) + p^2 \frac{T^2}{4} (z+1)^2}
$$

Then, multiplying out, combining like powers of $z$, and simplifying (assuming: $T = 10 ps$ and $\tau_d = 150 ps$):

$$
H(z) = -p \frac{0.98 \frac{T}{2} \left[ \left( 1 - p' \right) z^2 - 2 p' z - \left( 1 + p'\right) \right] - 0.005 p T^2 \left[ z^{-28} + 2 z^{-29} + z^{-30} \right]}{\left( 1 - p'\right)^2 z^2 - 2 \left( 1 + p'\right) \left( 1 - p'\right) z + \left( 1 + p'\right)^2}
$$

where:

$$
\begin{eqnarray}
p' &=& p \frac{T}{2}
\end{eqnarray}
$$


Finally, multiplying numerator and denominator by: $(1 - p')^{-2} z^{-2}$, gives us the standard normalized expanded form we've been aiming for:

$$
H(z) = -p \frac{0.98 \frac{T}{2} \left[ \left( 1 - p' \right)^{-1} - 2 p' (1 - p')^{-2} z^{-1} - \left( 1 + p'\right) (1 - p')^{-2} z^{-2} \right] - 0.005 p T^2 (1 - p')^{-2} \left[ z^{-30} + 2 z^{-31} + z^{-32} \right]}{1 - 2 \left( 1 + p'\right) \left( 1 - p'\right)^{-1} z^{-1} + \left( 1 + p'\right)^2 (1 - p')^{-2} z^{-2}}
$$

### Viewing the New Channel Impulse Response

Now, let's feed these coefficients into the `dimpulse()` function and see how we did...

````python
from scipy.signal import dimpulse

Npts = 8192
Ts   = 10e-12
t, f, w, s, z, fp = make_vecs(Ts, Npts)
p  = -2 * np.pi * 20e9
pp = p*Ts/2
b = list(-0.98*p*Ts/2 * np.array([(1-pp)**-1, -2*pp*(1-pp)**-2, -(1+pp)*(1-pp)**-2])) + 27*[0] + list(0.005*p**2*Ts**2*(1-pp)**-2*np.array([1, 2, 1]))
a = [1, -2*(1+pp)*(1-pp)**-1, (1+pp)**2*(1-pp)**-2] + 30*[0]
t, h = dimpulse((b, a, Ts))
_t = t - t[-1]/2
_h = np.fft.fftshift(h[0].ravel())
plt.figure(figsize=(10,4))
plt.plot(_t*1e9, _h)
plt.title("Channel Impulse Response - Correct Application of Bilinear Transform")
plt.xlabel("Time (ns)")
plt.show()
````

Note the following features of the plot above:

1. The channel impulse response is causal.
All ringing prior to the initial rising edge is gone.

1. The reflected pulse is in the proper location: 300 ps after the main pulse.

1. The FWHM of the main pulse is correct at approx.: $\frac{1}{20 GHz} = 50 ps$.

1. The ratio of reflection delay to main pulse FWHM is now correct at approx. 6.

### Checking the New Channel Frequency Response

Let's check the frequency response of our transformed digital system.

````python
from scipy.signal import freqz

_, Hz = freqz(b, a, worN=fp, fs=1/Ts)

plt.figure(figsize=(10,4))
plt.subplot(121)
plt.semilogx(fp/1e9, 20*np.log10(np.abs(Hz[:len(fp)])))
plt.title("Frequency Response - Magnitude")
plt.xlabel("Frequency (GHz)")
plt.ylabel("Mag.(H) (dB)")
plt.axis(ymin=-30)
plt.grid()

plt.subplot(122)
plt.semilogx(fp/1e9, np.unwrap(np.angle(Hz[:len(fp)])) *180/np.pi)
plt.title("Frequency Response - Phase")
plt.xlabel("Frequency (GHz)")
plt.ylabel("Phase(H) (deg.)")
plt.grid()

plt.show()
````

We see that both the -3dB magnitude and -45 deg. phase points are correctly located at 20 GHz, our chosen filter bandwidth.

In most Touchstone data sets we'll be working with (i.e. - those characterizing channels suitable for data transmission at 100s of Gbps) the data simply stop well before the channel response has come anywhere near to zero.
Does the Bilinear Transform solve this problem for us?

Note that at the point where the Touchstone data ends the insertion loss is still just above -10 dB.

### Vector fitting differential insertion loss

We're interested in reconstructing the differential throughput impulse response.
And, in order to apply the _Bilinear Transform_, we'll have to express the differential insertion loss as a polynomial expression in the s-domain.
So, as a first step, let's see how well we can vector fit this particular differential insertion loss.

````python
ntwk_mm = ntwk.copy()
ntwk_mm.renumber([0, 1, 2, 3], [0, 2, 1, 3])
ntwk_mm.se2gmm(p=2)
ntwk_dd = ntwk_mm.subnetwork([0, 1])
ntwk_dd.s21.plot_s_db(show_legend=False)
plt.title("Differential Insertion Loss for My Original Touchstone Data")
plt.grid()
````

Note that in the differential case the limitation is even worse: the Touchstone data end with the IL still just above -5 dB!

````python
vf = rf.VectorFitting(ntwk_dd.s21)
vf.vector_fit(n_poles_real=10, n_poles_cmplx=40)
vf.plot_s_db()
````

That looks pretty good except for that slight up-tick at the right edge of the plot, which is going to cause trouble later on.
(See Sec. 7.5 - _Using Just the Vector Fitting Results_.)

Let's check the quality of the fit quantitatively.

````python
_f = ntwk_dd.f
sig = ntwk_dd.s21.s[:, 0, 0]
Hs = vf.get_model_response(0, 0, _f)
mag_err = np.abs(Hs) - np.abs(sig)
rmse = np.sqrt(np.mean(mag_err**2))
rmss = np.sqrt(np.mean(np.abs(sig)**2))
print(f"Average fit SNR: {20 * np.log10(rmss / rmse):.0f} dB")
````

Okay, that satisfies Prof. Morales' requirement of: >40dB.

Let's transform the poles and see what the resultant discrete time unit impulse response looks like.

### Impulse Response Reconstruction, via _Bilinear Transform_

<div class="alert-info">
    At this point, you may be panicking, because you're envisioning us having to manually transform the 50 poles in the vector fitting result.
    And that would, indeed, render this approach completely untenable for practical engineering work.
    But, as luck would have it, all the common mathematical modeling packages provide a function for doing this polynomial ratio transformation from the s to the z-domain.
    In <em>NumPy/SciPy</em> it's called: <code>bilinear()</code>. <p>
    They also give us a function for converting from the <em>partial fractions</em> (i.e. - pole/residue) form output by the vector fitter to the <em>expanded polynomial</em> form expected by the other functions we're using.
    In <em>NumPy/SciPy</em> it's called: <code>invres()</code>.
</div>

Here are the _docstrings_ for those two functions:

````python
from scipy.signal import invres, bilinear, dimpulse  # , impulse, invresz, freqs, freqz, 

print("bilinear():")
print(bilinear.__doc__)
print("\ninvres():\n")
print(invres.__doc__)
````

Let's use the two functions above to apply the Bilinear Transform and see what the reconstructed impulse response of our discrete time system looks like.

````python
ba, aa = invres(vf.residues[0], vf.poles, vf.proportional_coeff[0])
print(f"ba: {ba}")
print(f"aa: {aa}")
````

And we immediately see the difficulty with trying to apply the BT to real World Touchstone channel data: it's not always possible to convert from the partial fractions form output by the vector fitter to the expanded polynomial form needed by the `bilinear()` function.

<div class="alert-info">
    There's actually been quite a bit written about this practical problem.
    One of the best papers I've found on this subject is 
    <em>Rational Function Computing with Poles and Residues</em> by
    Richard J. Fateman of U.C. Berkeley (Dec. 24, 2010).
</div>
    

### Using a Scaled Frequency Space

It appears, above, that we're just running out of room, wrt/ what a Python `float` can handle.
Might we succeed if we pre-scaled the poles by some common factor?
Let's find out...

````python
FSCALE = 1e-13  # Change the units of the poles from Hz to THz.
ntwk_dd_scaled = ntwk_dd.copy()
ntwk_dd_scaled.f *= FSCALE
vf_scaled = rf.VectorFitting(ntwk_dd_scaled.s21)
vf_scaled.vector_fit(n_poles_real=10, n_poles_cmplx=40)
vf_scaled.plot_s_db()
````

Okay, great! It looks the same except for the scaled horizontal axis, which is what we wanted.  
Now, do we have better luck trying to convert to expanded polynomial form?

````python
r = vf_scaled.residues[0]
p = vf_scaled.poles
k = vf_scaled.proportional_coeff[0]
ba_scaled, aa_scaled = invres(r, p, k, tol=1e-9)
print(f"ba_scaled: {ba_scaled}")
print(f"aa_scaled: {aa_scaled}")
````

Yes! that got rid of all the `NaN`s; great!  
What does the resultant (analog) frequency response look like?

````python
from scipy.signal import freqs

w_scaled, H_scaled = freqs(ba_scaled, aa_scaled)
f_scaled = w_scaled / (2 * np.pi)
H_scaled /= np.abs(H_scaled[0])  # Normalize for unity d.c. gain.
plt.figure(figsize=(10,4))
plt.semilogx(f_scaled, 20 * np.log10(np.abs(H_scaled)), label="Expanded Polynomial")
plt.semilogx(f_scaled, 20 * np.log10(np.abs(vf_scaled.get_model_response(0, 0, f_scaled))), label="Partial Fractions")
plt.title("Frequency Response of Scaled Transform - Mag.")
plt.xlabel("Frequency (Hz)")
plt.ylabel("|H_scaled(f)| (dB)")
plt.grid()
plt.legend()
plt.show()
````

Oof! I think we're way off track, even below 35 GHz (35 Hz in the scaled horizontal axis, above).  
It would appear that the `invres()` function is failing us, despite our attempt to accommodate it, via frequency scaling.

What happens if we run `aa_scaled` and `ba_scaled` thru `residue()`?  
We should get back our original partial fractions representation.  
Do we?

````python
from scipy.signal import residue

_r, _p, _k = residue(ba_scaled, aa_scaled)

# Check new vector lengths.
assert len(_r) == len(r)
assert len(_p) == len(p)
assert len(_k) == 0 and k == 0 or len(_k) == len(k)

# Check new vector contents.
r_diff = np.abs(_r - r)
p_diff = np.abs(_p - p)
if len(_k):
    k_diff = np.abs(_k - k)
    print(f"Maximum differences normalized to means (r, p, k): ({max(r_diff) / np.mean(r_diff):.3f}, {max(p_diff) / np.mean(p_diff):.3f}, {max(k_diff) / np.mean(k_diff):.3f})")
else:
    print(f"Maximum differences normalized to means (r, p, k): ({max(r_diff) / np.mean(r_diff):.3f}, {max(p_diff) / np.mean(p_diff):.3f}, 0)")
````

Hmmm, max. differences greater than twice the average value for both the residues and poles.  
That's probably not a good sign.

### Using Just the Vector Fitting Output

Well, let's not give up on the vector fit, just because we can't use it in a _Bilinear Transform_ (BT).
We can still calculate the frequency response from the vector fitting result and take the `ifft()` of that, to recover the impulse response.

**Note:** In this case, it's okay to use the `ifft()` function, since we won't be distorting the frequencies with the BT.

**Note:** We use the `irfft()` function to recover the windowed impulse response, below. This is just a convenience allowing us to omit the conjugate responses from the input vector without causing the sort of erroneous output that would ordinarily come from the `ifft()` function when we do this.

````python
Ts   = 10e-12  # 100 GHz sampling rate
Npts = 8192
print(f"Ts = {Ts}")
print(f"Npts = {Npts}")
t, f, w, s, z, fp = make_vecs(Ts, Npts)

# Calculate windowed IR, for reference.
# Generate manually constructed frequency domain data vector.
# Form frequency vector.
_f   = ntwk_dd.s21.f
fmin = _f[0]
if(fmin == 0):    # If the d.c. point was included in the Touchstone file...
    fmin = _f[1]  # then the next element contains the real fmin.
fmax = _f[-1]
Hdc  = 1          # Assume unity d.c. throughput.
HNyq = ntwk_dd.s21.s[-1, 0, 0]
_f   = np.arange(fmin, fmax, fmin)    # Enforcing uniform frequency steps.
F    = rf.Frequency.from_f(_f)  # skrf.Frequency.from_f() recently changed to use MKS units.
_f   = np.append(np.insert(_f, 0, 0), fmax)  # Adding back d.c. and Nyquist.

# Form impulse response from frequency response.
Ht  = ntwk_dd.s21.interpolate(F).s[:, 0, 0]
Ht  = np.append(np.insert(Ht, 0, Hdc), HNyq)  # `irfft()` expects the extra, Nyquist, component.
N   = len(Ht)
ix  = np.arange(N)
win = 0.5 * (np.cos(np.pi * ix / N) + 1)
Hw  = Ht * win  # windowed version of insertion loss data
hw  = np.fft.irfft(Hw)
_t  = np.array([n * 1/(2*fmax) for n in range(len(hw))])

# Calculate vector fit IR.
Hs = vf.get_model_response(0, 0, f)
hs = np.fft.ifft(Hs)

# Plot the frequency responses.
plt.figure(figsize=(10,4))
plt.semilogx(fp/1e9, 20*np.log10(np.abs(Hs[:len(fp)])), label="Vector Fit")
plt.semilogx(_f/1e9, 20*np.log10(np.abs(Hw)),           label="Windowed")
plt.title("Channel Frequency Responses")
plt.xlabel("Frequency (GHz)")
plt.ylabel("|H(f)|")
# plt.axis(xmin=-0.1, xmax=1.6)
plt.legend()
plt.show()

# Plot the impulse responses.
t_shift = t - max(t)/2
plt.figure(figsize=(10,4))
plt.plot(t_shift*1e9, np.fft.fftshift(hs), label="Vector Fit")
plt.plot(_t*1e9,      hw,                  label="Windowed")
plt.title("Channel Impulse Responses")
plt.xlabel("Time (ns)")
plt.ylabel("h(t)")
plt.axis(xmin=-0.1, xmax=1.6)
plt.legend()
plt.show()
````

Hmmm, my windowed approach is looking pretty good right about now. ;-)

**Note:** The vector fit is producing strange behavior beyond the 35 GHz limit of the Touchstone data, which I'm sure is not physically correct.
It looks like its extrapolation algorithm is getting confused.
This is, no doubt, partially responsible for the odd looking impulse response.

Oh, I see: take another look at the fitting results plot, above, and notice that little up-tick in the fitted curve near the right edge of the plot.
That must be the culprit.

**This is a wonderful example of the fragility of the vector fitting approach.**
